```python
latitude = 12.9716
longitude = 77.5946
search_query = 'bar'
radius = 1000
LIMIT = 500
```

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
url
```

```python
results = requests.get(url).json()['response']['venues']
#a = []
#a.append([b['name'] for b in results])
#a
results
#dataframe = json_normalize(results)
#dataframe.head()
```

```python
venues_list = []
for items in results:
    try:
        v_name = items['name']
    except:
        v_name = ''
    try:
        v_add = items['location']['formattedAddress'][0]
    except:
        v_add = ''
    try:
        v_c_name = items['categories'][0]['name']
    except:
        v_c_name = ''

    venues_list.append([(items['id'],v_name, v_add, v_c_name)])
                     
#venues_list
```

```python
nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
```

```python
nearby_venues.columns= ['ID', 'Venue Name', 'Address', 'Venue Category']
```

```python
def find_rating(venue_id):
    CLIENT_ID = os.environ['CLIENT_ID'] # your Foursquare ID
    CLIENT_SECRET = os.environ['CLIENT_SECRET']  # your Foursquare Secret
    VERSION = '20200411' # Foursquare API version
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID, CLIENT_SECRET, VERSION)
    result = requests.get(url).json()
    #print(result['response']['venue'].keys())
    #return result['response']['venue']
    try:
        rating = result['response']['venue']['rating']
    except:
        rating = 'This venue has not been rated yet.'
    return rating
    
```
